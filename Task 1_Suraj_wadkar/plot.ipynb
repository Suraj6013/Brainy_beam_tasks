# ðŸ“Š Multiple Y-Axes Sharing a Common X-Axis Plot ðŸ“ˆ

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
days = np.arange(1, 31)  # Days of the month
hours_of_sunshine = np.linspace(3, 10, num=30)  # Gradually increasing hours of sunshine from 3 to 10 hours
energy_production = np.random.uniform(150, 500, size=30)  # Random energy production between 150 and 500 kWh

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the hours of sunshine on the primary y-axis
ax1.plot(days, hours_of_sunshine, 'b-', label='Hours of Sunshine')
ax1.set_xlabel('Day of the Month')
ax1.set_ylabel('Hours of Sunshine', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_ylim(0, 12)  # Set the y-axis limits for hours of sunshine
ax1.grid(True)

# Create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

# Plot the energy production on the secondary y-axis
ax2.plot(days, energy_production, 'r-', label='Energy Production (kWh)')
ax2.set_ylabel('Energy Production (kWh)', color='r')
ax2.tick_params(axis='y', labelcolor='r')
ax2.set_ylim(0, 600)  # Set the y-axis limits for energy production

# Set the same x-axis limits for both axes
ax1.set_xlim(0, 31)
ax2.set_xlim(0, 31)

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Add a title and show the plot
plt.title('Hours of Sunshine vs Energy Production Over a Month')
fig.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

# Hours of Sunshine vs Energy Production Plot

This plot visualizes the relationship between daily hours of sunshine and solar panel energy production over a month. So two y-axes share a common x-axis here
## Data Description
- **Days of the Month**: The x-axis represents the days of the month (1 to 30).
- **Hours of Sunshine**: The primary y-axis (left) shows the hours of sunshine per day, gradually increasing from 5 to 10 hours.
- **Energy Production**: The secondary y-axis (right) shows the energy production in kilowatt-hours (kWh), with random values between 200 and 500 kWh.

## Plot Details
- The blue line represents the hours of sunshine.
- The red line represents the energy production.
- The first y-axis is labeled "Hours of Sunshine" and ranges from 0 to 12 hours.
- The second y-axis is labeled "Energy Production (kWh)" and shows the energy production values.
