```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
days = np.arange(1, 31)  # Days of the month
temperature = np.linspace(30, 35, num=30)  # Gradually increasing temperatures from 30 to 35 degrees Celsius
ice_cream_sales = np.random.uniform(50, 150, size=30)  # Random ice cream sales between 50 and 150 units

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the temperature on the primary y-axis
ax1.plot(days, temperature, 'b-', label='Temperature (°C)')
ax1.set_xlabel('Day of the Month')
ax1.set_ylabel('Temperature (°C)', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.grid(True)

# Create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

# Plot the ice cream sales on the secondary y-axis
ax2.plot(days, ice_cream_sales, 'r-', label='Ice Cream Sales (units)')
ax2.set_ylabel('Ice Cream Sales (units)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Add a title and show the plot
plt.title('Temperature vs Ice Cream Sales Over a Month')
fig.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
